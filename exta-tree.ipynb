```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.ensemble import ExtraTreesClassifier 
```

```python
df = pd.read_csv("UCP-Plus.csv")
df = df.drop(["Real_P20"], axis=1)
X = df.drop(["Effort"], axis=1)
y = df["Effort"]
```

```python
ET = ExtraTreesClassifier(n_estimators=5, criterion="entropy", max_features=2)
ET.fit(X,y)
```

```python
FI = ET.feature_importances_
```

```python
FIN = np.std([tree.feature_importances_ for tree in ET.estimators_], axis = 0)
FIN2 = FIN
FIN = sorted(FIN)
```

```python
fig = plt.gcf()
plt.rcParams.update({'font.size': 16})
fig.set_size_inches(28, 10.5)
fig.savefig('test2.png', dpi=100)

plt.bar(X.columns, FIN) 
plt.xlabel('Feature Labels') 
plt.ylabel('Feature Importances') 
plt.title('Comparison of Different Feature Importances in the UCP Dataset')
plt.show() 
```
